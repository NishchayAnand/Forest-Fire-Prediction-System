```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
import tensorflow as tf
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
df = pd.read_excel("forestfires_with_FWI.xlsx")
```

```python
df.head()
```

```python
df.shape
```

```python
input_set = df.drop(["X","Y","area"], axis = 1)
input_set.shape
```

```python
month_index={"jan":1, "feb":2, "mar":3, "apr":4, "may":5, "jun":6, "jul":7, "aug":8, "sep":9, "oct":10, "nov":11, "dec":12}
week_index = {'sun':1,'mon':2,'tue':3,'wed':4,'thu':5,'fri':6,'sat':7}
input_set["month"].replace(month_index, inplace=True)
input_set["day"].replace(week_index, inplace=True)
```

```python
input_set.head()
```

```python
input_set.columns=["Month","Day","FFMCo","DMCo","DCo","ISI","Temp","Rhum","Wind","Rain","FWI"]
input_set = input_set.round()
input_set.head()
```

```python
input_set.to_excel("Fire_Weather_Data_1.xlsx", index = False)
```
