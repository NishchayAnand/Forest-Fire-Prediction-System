```python
import pandas as pd
```

```python
df = pd.read_excel("Fire_Weather_Data.xlsx")
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
arr = []
```

```python
for i in range(df.shape[0]):
    arr1 = np.array(df.iloc[i,:])
    arr.append(arr1)
```

```python
labels = ["JAN","FEB","MAR","APR","MAY","JUN","JUL","AUG","SEP","OCT","NOV","DEC"]
```

```python
plt.figure(figsize=[10,6])
for i in range(len(arr)):
    plt.plot(range(len(arr[i])),arr[i],label=labels[i])
plt.legend(loc='best')
plt.ylabel('Amount of Rainfall (in mm)', fontsize="14")
plt.xlabel('Years (1901 - 2015)', fontsize="14") 
plt.show()
```
